Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves two key hypotheses: the null hypothesis and the alternative hypothesis. The null hypothesis represents no effect or no difference, acting as the default assumption, while the alternative hypothesis contradicts this and is what we aim to support. Hypothesis testing seeks to determine if there is enough evidence to reject the null hypothesis in favor of the alternative. Errors can occur, such as Type I errors or Type II errors. A significance level, often 0.05 or 0.01, helps set the threshold for making this decision, balancing the risk of errors.

A p-value is a measure used in hypothesis testing to help determine the strength of evidence against the null hypothesis. Assuming the null hypothesis is true, it represents the probability of obtaining results as extreme as, or more extreme than, what was observed in sample data. 
Suppose the p-value is small (typically less than the significance level, often 0.05). In that case, it suggests that the observed results are unlikely under the null hypothesis, providing evidence to reject it in favor of the alternative hypothesis. A more significant p-value means that the observed results are consistent with the null hypothesis, and there is insufficient evidence to reject it. Notably, a p-value does not provide a definitive answer or proof but indicates whether the data supports rejecting or retaining the null hypothesis.

The statement "it is important to note that outcomes of tests refer to the population parameter, rather than the sample statistic" highlights that in hypothesis testing, our main objective is to make inferences about the entire population from which the sample is drawn, rather than focusing solely on the characteristics of the sample itself.

During hypothesis testing, we assume the null hypothesis is accurate. The null hypothesis makes a claim about the population parameter, such as the population mean, suggesting it is equal to a specific value. When we collect a sample, we observe individual data points and compute the sample average. The sample mean is used as an estimate of the population mean. Still, it is essential to remember that this sample mean is influenced by random variation and might not perfectly represent the actual population mean.

The purpose of hypothesis testing is to use the information from our sample to assess whether there is enough evidence to reject the null hypothesis about the population parameter. For instance, if our sample average significantly differs from the hypothesized value, we might conclude that the population mean will likely not equal the hypothesized value. In this context, the focus remains on deciding the population mean, not merely describing the sample statistic.

In simple terms, although we use data from the sample to perform calculations, our end goal is to reach conclusions that apply to the entire population. The distinction between population parameters and sample statistics is crucial because it helps clarify that the results of hypothesis testing are intended to reflect the broader population characteristics, not just the sample used in the study. Thus, when conducting tests, we are interested in the sample data about the whole population, so our conclusions are centered around population parameters.

Summary of Exchanges on p-Value and Sampling Distribution Concepts

In our discussion, we explored the concept of p-values and their relationship to the sampling distribution in hypothesis testing. Here are the key points covered:

p-Value and its Interpretation:

The p-value is the probability of observing a test statistic at least as extreme as the one obtained, assuming the null hypothesis is true.
A smaller p-value suggests that the observed test statistic is unlikely under the null hypothesis, providing stronger evidence against it.
Sampling Distribution Under the Null Hypothesis:

The sampling distribution represents the range of possible values of the test statistic we would expect under the assumption that the null hypothesis is true.
This distribution is typically centered around the hypothesized value, with most values clustering near the center due to random variation.
Extreme Values and p-Value:

Extreme values in the sampling distribution lie in the tails, which means they are far from the expected center of the distribution.
The p-value measures how likely it is to observe such extreme values under the null hypothesis.
A small p-value means the observed value is in an extreme region of the sampling distribution, suggesting that it is unlikely to have occurred if the null hypothesis were true.
Evidence Against the Null Hypothesis:

A small p-value indicates that the observed test statistic is far from what we would expect under the null hypothesis.
This provides evidence against the null hypothesis, as the observed data is inconsistent with the assumption that it resulted purely from random chance.
Consequently, a small p-value often leads to rejecting the null hypothesis in favor of the alternative hypothesis, suggesting the presence of a statistically significant effect.
Role of the Sampling Distribution:

The sampling distribution of the test statistic helps determine whether the observed result is typical or unusual under the null hypothesis.
If the test statistic lies in the tail(s) of the sampling distribution (resulting in a small p-value), it implies that the null hypothesis is unlikely to explain the observed data, thereby undermining the null hypothesis.

Why One-Sided Tests Often Yield Smaller p-Values:

The p-value tends to be smaller in a one-sided test compared to a two-sided one because the critical region is concentrated in one direction rather than split between both tails. This makes the p-value smaller since less of the area is considered.

Interpretation Differences Between One-Sided and Two-Sided Tests:

We discussed that the interpretation of results differs between these tests. A one-sided test is specific to one direction, making it more powerful in detecting that effect but less flexible. A two-sided test is more general, testing for any difference, which makes it more conservative but suitable when you lack a strong directional hypothesis.

Applicability and Risks:

A one-sided test is useful when there is a specific hypothesis about the direction of an effect, but there is a risk of missing an effect in the opposite direction. A two-sided test avoids this risk by considering both directions and is often used when you don't have a strong prior belief about the effect's direction.

```python
import numpy as np
from scipy.stats import norm

# Sample data
n_cups = 8  # Total number of cups
n_correct = 5  # Number of correct guesses made by the student
confidence_level = 0.95

# Calculate sample proportion
p_hat = n_correct / n_cups

# Calculate standard error
se = np.sqrt(p_hat * (1 - p_hat) / n_cups)

# Find the critical value for the desired confidence level
z_star = norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z_star * se

# Calculate the confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

print(f"95% Confidence Interval for the proportion of correct guesses: ({lower_bound:.2f}, {upper_bound:.2f})")
```

To assess the validity of the null hypothesis, we used simulation to generate a sampling distribution under the assumption that the null hypothesis is true. We performed a bootstrap simulation, repeatedly sampling 80 students, each with a 50% chance of answering correctly. We then compared the observed value (49 correct responses) to this simulated distribution to calculate the p-value, which indicates how likely it is to observe a result as extreme as or more extreme than the one we obtained if the null hypothesis is true.

Based on the simulation results, we determined whether the p-value was sufficiently low (typically less than 0.05) to reject the null hypothesis. A small p-value would suggest that the outcome of 49 correct responses is unlikely to be due to random guessing, providing evidence against the null hypothesis. Python Code for Bootstrap Simulation:
Here's the Python code that you can use to simulate the sampling distribution and calculate the p-value for Fisher's Tea Experiment:

```python
import numpy as np
import matplotlib.pyplot as plt

# Setting a seed for reproducibility
np.random.seed(42)

# Parameters for the experiment
n_students = 80
n_correct = 49
n_simulations = 10000

# Simulating the distribution under H0 (null hypothesis)
# Assuming that the probability of getting it correct is 0.5 (random guessing)
simulated_correct = np.random.binomial(n_students, 0.5, n_simulations)

# Calculating the p-value: probability of getting a result as extreme or more extreme than observed
p_value = np.mean(simulated_correct >= n_correct)

# Printing p-value
print(f'P-value: {p_value}')

# Plotting the distribution of simulated outcomes
plt.hist(simulated_correct, bins=np.arange(25, 81) - 0.5, edgecolor='black')
plt.axvline(n_correct, color='red', linestyle='dashed', linewidth=2, label=f'Observed = {n_correct}')
plt.xlabel('Number of Correct Responses')
plt.ylabel('Frequency')
plt.title('Sampling Distribution under Null Hypothesis')
plt.legend()
plt.show()
```

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)
- [Scientists rise up against statistical significance](https://www.nature.com/articles/d41586-019-00857-9)
- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

1. Hypothesis testing is not a "mathematical proof"<br><br>

    1. We do not prove $H_0$ false, we instead give evidence against the $H_0$: "We reject the null hypothesis with a p-value of XYZ, meaning we have ABC evidence against the null hypothesis"
    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>

2. Implying that a "non-significant result" means there is "no effect" misleads an audience because this may in actual fact simply indicate that there was insufficient evidence to reject the null hypothesis. So this therefore overlooks the possibility of sample size limitations, or Type II errors (which means a test incorrectly concludes that there is no effect or difference when, in fact, there is one). 
    
> Similarly, analagously, a "significant result" used to reject the null hypothsis could alternatively be a Type I error (which means a test actually incorrectly rejected a null hypothesis when it was actually true)... we're only providing a measure of evidence against the null hypothesis... but the evidence could still incorrectly suggest the wrong conclusion... it really depends on how strong the evidence is...
>
> - all of which is why just interpreting p-values using the table above is a good idea...

The p-values used for hypothesis testing are contructed upone the assumptions of the null hypotheses they correspond to; but, null hypotheses are actually often presented in simple forms that routinely hide a lot of information that is implicitly used to construct the p-values. An alternative hypothesis that "the null hypothesis is false" doesn't say much... we should wonder, "how, specfically, is the null false?"
